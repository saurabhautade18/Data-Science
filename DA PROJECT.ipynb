# Covid-19 cases Analysis across globe


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
os.getcwd()
```

```python
data = pd.read_excel('country_wise_latest1.xlsx')
data
```

```python
plt.plot(data['Country/Region'],data['Deaths'],marker='o',linewidth=3,color='k');
plt.xlabel('Country')
plt.ylabel('Deaths')
plt.title('COVID-19 deaths Over Globe ')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.bar(data['Country/Region'],data['Active'],color='r');
plt.xlabel('Country')
plt.ylabel('Active Patients')
plt.title('COVID-19 Active patients Over Globe ')
plt.xticks(rotation=40)
plt.legend('active patients')

plt.show()
```

```python
sr=data[['Confirmed','Deaths']]
sr
```
